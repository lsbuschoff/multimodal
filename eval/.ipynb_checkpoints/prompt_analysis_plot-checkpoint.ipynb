```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pkg_resources
import seaborn as sns
```

```python
def plot_params():

    mpl.rcParams["axes.spines.right"] = False
    mpl.rcParams["axes.spines.top"] = False
    
    FIGWIDTH = 7.00697
    SMALL_SIZE = 8
    MEDIUM_SIZE = 10
    BIGGER_SIZE = 12
    
    #  gray for fuyu, purple for adapter, teal for otter, green for gpt-4v, orange for claude, red for humans
    model_colors = [
        "#8c92ac",
        "#624fe8",
        "#008080",
        "#50b990",
        "#da7756",
        "#9b443e",  
    ]
    
    cmap = [
        "#53665c",
        "#ccb3a0",
        "#f4b368",
        "#a08158",
        "#c87858",
        "#94b1d2",
        "#c5d9d8",
        "#e6b951",
        "#666666",
    ]
    sns.set_palette(cmap)
    
    plt.rc("font", size=SMALL_SIZE)  # controls default text sizes
    plt.rc("axes", titlesize=SMALL_SIZE)  # fontsize of the axes title
    plt.rc("axes", labelsize=SMALL_SIZE)  # fontsize of the x and y labels
    plt.rc("xtick", labelsize=SMALL_SIZE)  # fontsize of the tick labels
    plt.rc("ytick", labelsize=SMALL_SIZE)  # fontsize of the tick labels
    plt.rc("legend", fontsize=SMALL_SIZE)  # legend fontsize
    plt.rc("figure", titlesize=BIGGER_SIZE)  # fontsize of the figure title
    
    mpl.rcParams["figure.dpi"] = 300
    mpl.rcParams["savefig.dpi"] = 300

    return cmap, model_colors
```

### Task settings

```python
# Load data from our own collected .csv file
data = pd.read_csv("../data/NUC_RESULTS.csv").loc[:, :"OTTER"]

# Define subjects and conds
subjects = ["gpt4", "prev", "fuyu"]
conds = ["1_basic_number", "2_easy_number", "3_basic_cot", "4_easy_cot"]

# Limit data to only first experiment
data = data[data["Id"] == "a"]

# Print first 25 rows which show all responses to the first 8 experiments
data.head()
```

### Make arrays

```python
# Read output file
for cond_ind, cond in enumerate(conds):

    # Load output file for this condition
    out = pd.read_csv(f"results/prompt_analysis/{cond}_comb.csv", header=0)

    # Loop through models
    for model_ind, model in enumerate(subjects):
            
        # Re-format output
        answers = out[model]
        answers = answers.astype(float)

        # Store model answers and human mean answers for cost questions
        cost_inds = data["Type"] == "Cost"     
        num_seqs = sum(cost_inds)
        temp = np.zeros((num_seqs, 2))
        temp[:] = np.nan
        temp[:, 0] = answers[cost_inds]
        temp[:, 1] = data["Human"][cost_inds]
        temp = pd.DataFrame(temp, columns=["Model", "Human"])
        temp.to_csv(f"../dataframes/prompt_analysis/{model}_cost_{cond}.csv")

        # Reward correlation
        reward_inds = data["Type"] == "Reward"
        num_seqs = sum(reward_inds)
        temp = np.zeros((num_seqs, 2))
        temp[:] = np.nan
        temp[:, 0] = answers[reward_inds]
        temp[:, 1] = data["Human"][reward_inds]
        temp = pd.DataFrame(temp, columns=["Model", "Human"])
        temp.to_csv(f"../dataframes/prompt_analysis/{model}_reward_{cond}.csv")
```

```python
reward_lmer_coef = []
reward_lmer_conf = []
cost_lmer_coef = []
cost_lmer_conf = []

for cond in conds:

    # Load files
    reward_lmer = pd.read_csv(f'../dataframes/prompt_analysis/reward_lmer_{cond}.csv')
    cost_lmer = pd.read_csv(f'../dataframes/prompt_analysis/cost_lmer_{cond}.csv')

    # Save coefs
    reward_lmer_coef.append(reward_lmer.iloc[:,0].values)
    cost_lmer_coef.append(cost_lmer.iloc[:,0].values)

    # Save confs
    reward_lmer_conf.append(reward_lmer.iloc[:,1:].values.T)
    cost_lmer_conf.append(cost_lmer.iloc[:,1:].values.T)

reward_confs = np.hstack(reward_lmer_conf)
cost_confs = np.hstack(cost_lmer_conf)
```

### Load R correlation

```python
# Make array
arr = np.zeros((24, 4), dtype='object')  # Long format
arr[:, 0] = np.repeat(["cost", "reward"], 12)  # Returns 12x cost then 12x reward
arr[:, 1] = np.repeat(np.expand_dims(np.repeat(conds, 3), 0), 2, axis=0).flatten()  # Returns 3x (3x 0.25 then 3x 0.5 then 3x 0.85)
arr[:, 2] = np.repeat(np.expand_dims(subjects, 0), 8, 0).flatten()  # Returns 9x (0, 1, 2)
arr[:, 3] = np.hstack((np.array(cost_lmer_coef).flatten(), np.array(reward_lmer_coef).flatten()))  # Stack the answers for the models in long format
df = pd.DataFrame(arr, columns=["question", "cond", "model", "val"])  # Assign column names
```

```python
# Init plotting params
cmap, model_colors = plot_params()

# Prepare plot
num_subjects = len(subjects)
fig, axs = plt.subplots(1, 2, figsize=(8,3), sharey=False, tight_layout=True)  
r1 = np.arange(3)
barWidth = 0.2
r2 = r1 + barWidth
r3 = r2 + barWidth
r4 = r3 + barWidth

# Plot B
axs[0].bar(r1, df[(df.question == "cost") & (df.cond == conds[0])].val, yerr=np.array(cost_lmer_conf)[0, :, :],
           color=cmap[0], width=barWidth, edgecolor='white', label='Basic Number')
axs[0].bar(r2, df[(df.question == "cost") & (df.cond == conds[1])].val, yerr=np.array(cost_lmer_conf)[1, :, :], 
           color=cmap[1], width=barWidth, edgecolor='white', label='Simple Number')
axs[0].bar(r3, df[(df.question == "cost") & (df.cond == conds[2])].val, yerr=np.array(cost_lmer_conf)[2, :, :], 
           color=cmap[2], width=barWidth, edgecolor='white', label='Basic CoT')
axs[0].bar(r4, df[(df.question == "cost") & (df.cond == conds[3])].val, yerr=np.array(cost_lmer_conf)[3, :, :], 
           color=cmap[3], width=barWidth, edgecolor='white', label='Simple CoT')

axs[0].set_xticks(r1 + 1.5*barWidth)
axs[0].set_xticklabels(["GPT-4V", "GPT-4V\nPreview", "Fuyu"])
axs[0].set_title("Cost")
axs[0].set_title("B", fontweight='bold', loc='left')
axs[0].set_ylabel(f"Pseudo $r$ to humans")
axs[0].set_xlabel("")
axs[0].set_ylim(-1, 1)

# Plot B
axs[1].bar(r1, df[(df.question == "reward") & (df.cond == conds[0])].val, yerr=np.array(reward_lmer_conf)[0, :, :],
           color=cmap[0], width=barWidth, edgecolor='white', label='Basic Number')
axs[1].bar(r2, df[(df.question == "reward") & (df.cond == conds[1])].val, yerr=np.array(reward_lmer_conf)[1, :, :], 
           color=cmap[1], width=barWidth, edgecolor='white', label='Simple Number')
axs[1].bar(r3, df[(df.question == "reward") & (df.cond == conds[2])].val, yerr=np.array(reward_lmer_conf)[2, :, :], 
           color=cmap[2], width=barWidth, edgecolor='white', label='Basic CoT')
axs[1].bar(r4, df[(df.question == "reward") & (df.cond == conds[3])].val, yerr=np.array(reward_lmer_conf)[3, :, :], 
           color=cmap[3], width=barWidth, edgecolor='white', label='Simple CoT')

axs[1].set_xticks(r1 + 1.5*barWidth)
axs[1].set_xticklabels(["GPT-4V", "GPT-4V\nPreview", "Fuyu"])
axs[1].set_title("Reward")
axs[1].set_title("B", fontweight='bold', loc='left')
axs[1].set_ylabel(f"Pseudo $r$ to humans")
axs[1].set_xlabel("")
axs[1].set_ylim(-1, 1)

# Despine and set x tick labels afterwards
sns.despine(top=True, right=True)
sns.despine(offset=5, trim=True)
plt.legend()
handles, labels = axs[1].get_legend_handles_labels()
axs[1].legend(handles, ['Basic prompt\nSingle number', 'Easy prompt\nSingle number', 
                        'Basic prompt\nStep by step','Easy prompt\nStep by step'],
              loc="upper left", bbox_to_anchor=(1.1, 0.9), ncol=1, title=None, frameon=False, labelspacing = 2)
plt.savefig("../figures/prompt_analysis_main.pdf", dpi=300)
```
