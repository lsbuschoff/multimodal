```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, kendalltau, fisher_exact, ttest_ind, sem, binom
from helpers import plot_params
from natsort import natsorted, natsort_keygen
```

## Load data

```python
# Define subjects
subjects = ["FUYU", "OTTER", "ADAPTER", "GPT-4V", "CLAUDE-3"]

# Load data from our own collected .csv file
data = pd.read_csv("data/ASTRONAUT_RESULTS.csv").loc[:, :subjects[-1]]

# Print first five rows
data.head()
```

### 1. Analyze match to ground truth for color test

```python
# Load data from our own collected .csv file
descriptives = pd.read_csv("data/ASTRONAUT_DESCRIPTIVES.csv")
descriptives = descriptives[descriptives["Experiment"] == "a"]
descriptives.head()

# Get percentage of correct background color over all three experiments
colors = np.zeros(len(subjects))
colors_errs = np.zeros(len(subjects))
num_images = len(descriptives.Image.unique())

for ind, subject in enumerate(subjects):
    correct = sum(descriptives[descriptives["Question"] == "Color"][subject].isin(["purple", "pink"]))
    colors[ind] = (correct / num_images) * 100
    colors_errs[ind] = binom.std(num_images, correct / num_images)
    print(f"Color, {subject}: {colors[ind]:.4f} %")
```

### 2. Analyze visual understanding test with number of containers in image

```python
# Get percentage of correct background color over all three experiments
numbers = np.zeros(len(subjects))
numbers_errs = np.zeros(len(subjects))

for ind, subject in enumerate(subjects):
    correct = sum(descriptives[descriptives["Question"] == "Number"][subject] == descriptives[descriptives["Question"] == "Number"]["GT"])
    numbers[ind] = (correct / num_images) * 100
    numbers_errs[ind] = binom.std(num_images, correct / num_images)
    print(f"Number, {subject}: {numbers[ind]:.4f} %")
```

### 3. Regression with human data for cost condition

```python
# Save human answers next to model answers for all models
# Output shape here is: (Number of humans * Number of test images, 3) 
# With 3 being: Model answers repeated number of humans times, All human answers, Human index
cost = data[data["Type"] == "Cost"]
num_seqs = len(cost)

for model in (subjects):

    # Loop through all human subjects
    temp = np.zeros((num_seqs, 2))
    temp[:] = np.nan

    # Store model answers and human mean answers
    temp[:, 0] = cost[model]
    temp[:, 1] = cost["Human"]
    print(f"{model}, {spearmanr(temp[:, 0], temp[:, 1], nan_policy="omit").statistic:.4f}")

    # Save to read in R after
    temp = pd.DataFrame(temp, columns=["Model", "Human"])
    temp.to_csv(f"dataframes/04_astronaut/{model}_cost.csv")
```

```python
# Load
cost_lmer = pd.read_csv('dataframes/04_astronaut/04_cost_lmer.csv')

# Separate coefficient and re-format CI
cost_lmer_coef = cost_lmer.iloc[:,0].values
cost_lmer_conf = [(cost_lmer.iloc[:,0] - cost_lmer.iloc[:,1]).values,
                  (cost_lmer.iloc[:,2] - cost_lmer.iloc[:,0]).values]
```

### 4. Regression with human data for reward condition

```python
# Save human answers next to model answers for all models
# Output shape here is: (Number of humans * Number of test images, 3) 
# With 3 being: Model answers repeated number of humans times, All human answers, Human index
reward = data[data["Type"] == "Reward"]
num_seqs = len(reward)

for model in (subjects):

    # Loop through all human subjects
    temp = np.zeros((num_seqs, 2))
    temp[:] = np.nan

    # Store model answers and human mean answers
    temp[:, 0] = reward[model]
    temp[:, 1] = reward["Human"]
    print(f"{model}, {spearmanr(temp[:, 0], temp[:, 1], nan_policy="omit").statistic:.4f}")

    # Save to read in R after
    temp = pd.DataFrame(temp, columns=["Model", "Human"])
    temp.to_csv(f"dataframes/04_astronaut/{model}_reward.csv")
```

```python
# Load
reward_lmer = pd.read_csv('dataframes/04_astronaut/04_reward_lmer.csv')

# Separate coefficient and re-format CI
reward_lmer_coef = reward_lmer.iloc[:,0].values
reward_lmer_conf = [(reward_lmer.iloc[:,0] - reward_lmer.iloc[:,1]).values,
                    (reward_lmer.iloc[:,2] - reward_lmer.iloc[:,0]).values]
```

## Make main plot
I imagine a plot with 4 panels, single row with four columns,  for every study
- A: Barplots model comparison for background color task. Bars show percentage correct.
- B: Counting model comparison as in A
- C: Reasoning ground truth comparison as in A
- D: Match to humans,  probably via regression coefficient.

```python
# Init plotting params
model_colors = plot_params()

# Prepare plot
num_subjects = len(subjects)
fig, axs = plt.subplots(1, 4, figsize=(8,3), sharey=False, tight_layout=True)  

# Plot A
axs[0].bar(np.arange(num_subjects), colors, color=model_colors, width=0.6)
axs[0].errorbar(np.arange(num_subjects), colors, yerr=colors_errs, ls="none", ecolor="gray")
axs[0].set_title("Background color")
axs[0].set_title("A", fontweight='bold', loc='left')
axs[0].set_ylabel("Percentage correct")
axs[0].set_ylim(0, 100)
axs[0].set_xticks(np.arange(num_subjects))

# Plot B
axs[1].bar(np.arange(num_subjects), numbers, color=model_colors, width=0.6)
axs[1].errorbar(np.arange(num_subjects), numbers, yerr=numbers_errs, ls="none", ecolor="gray")
axs[1].set_title("Number of boxes")
axs[1].set_title("B", fontweight='bold', loc='left')
axs[1].set_ylabel("Percentage correct")
axs[1].set_ylim(0, 100)
axs[1].set_xticks(np.arange(num_subjects))

# Plot C
axs[2].bar(np.arange(num_subjects), cost_lmer_coef, color=model_colors, width=0.6)
axs[2].errorbar(np.arange(num_subjects), cost_lmer_coef, yerr=cost_lmer_conf, ls="none", ecolor="gray")
axs[2].set_title("Cost")
axs[2].set_title("C", fontweight='bold', loc='left')
axs[2].set_ylabel("Regression to humans")
axs[2].set_ylim(-1, 1)
#axs[2].set_yticks(np.arange(-1, 1.1, 0.5))
axs[2].set_xticks(np.arange(num_subjects))

# Plot C
axs[3].bar(np.arange(num_subjects), reward_lmer_coef, color=model_colors, width=0.6)
axs[3].errorbar(np.arange(num_subjects), reward_lmer_coef, yerr=reward_lmer_conf, ls="none", ecolor="gray")
axs[3].set_title("Reward")
axs[3].set_title("D", fontweight='bold', loc='left')
axs[3].set_ylabel("Regression to humans")
axs[3].set_ylim(-1, 1)
#axs[3].set_yticks(np.arange(-1, 1.1, 0.5))
axs[3].set_xticks(np.arange(num_subjects))

# Despine and set x tick labels afterwards
sns.despine(offset=5, trim=True)
axs[0].set_xticklabels(subjects, rotation=90)
axs[1].set_xticklabels(subjects, rotation=90)
axs[2].set_xticklabels(subjects, rotation=90)
axs[3].set_xticklabels(subjects, rotation=90)
plt.savefig("figures/04_astronaut_main.pdf", dpi=300)
```
